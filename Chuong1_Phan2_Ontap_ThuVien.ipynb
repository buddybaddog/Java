## 1. Ôn tập Numpy
---
Thư viện Numpy là thư viện số, chuyên xử lý dữ liệu dạng mảng nhiều chiều.

```python
#gọi thư viện
import numpy as np

#tạo một ma trận các số ngẫu nhiên trong khoảng từ 1-100

matran_A = np.random.randint(1, 100, (10,15))
print(matran_A)
print('Kích thước của ma trận:', matran_A.shape)
```

```python
#Tách ma trận con số 1: lấy tất cả các hàng và 10 cột đầu tiên
matran_A1 = matran_A[:,0:10]
matran_A1
```

```python
#Tách ma trận con số 2: lấy tất cả các hàng và 5 cột cuối cùng
matran_A2 = matran_A[:,10:]
matran_A2
```

```python
#cách truy cập dữ liệu trong Numpy
print(matran_A1.shape)
print(matran_A2.shape)
```

```python
hangdau = matran_A[0,:]
hangdau
```

```python
#cách truy cập dữ liệu trong Numpy
hangcuoi = matran_A[-1,:]
hangcuoi
```

## 2. Ôn tập Matplotlib
---
Thư viện matplotlib là thư viện chuyên sử dụng để trực quan hóa - vẽ các biểu đồ

```python
#gọi thư viện
import matplotlib.pyplot as plt

ten_khoa_x = ['K64', 'K65','K66','K67','K68']
so_hs_y = [450, 570, 1520, 820, 200]

#Vẽ biểu đồ
```

```python
#Vẽ biểu đồ dạng cột thẳng đứng:
plt.bar(ten_khoa_x, #dữ liệu trục x
        so_hs_y, #dữ liệu trục y
        color='brown') #màu biểu đồ
plt.title('BIỂU ĐỒ THỂ HIỆN SỐ LƯỢNG SV THEO TỪNG KHÓA')
plt.grid(ls='--')
plt.xlabel('Khóa')
plt.ylabel('Số lượng')
plt.show()
```

```python
#Vẽ biểu đồ dạng cột nằm ngang:
plt.barh(ten_khoa_x, #dữ liệu trục x
        so_hs_y, #dữ liệu trục y
        color='black') #màu biểu đồ
plt.title('BIỂU ĐỒ THỂ HIỆN SỐ LƯỢNG SV THEO TỪNG KHÓA')
plt.grid(ls='--')
plt.xlabel('Số lượng')
plt.ylabel('Khóa')
plt.show()
```

```python
#vẽ biểu đồ dạng đường
plt.plot(ten_khoa_x,
        so_hs_y,
        'b-o',ms=10)

plt.title('BIỂU ĐỒ THỂ HIỆN SỐ LƯỢNG SV THEO TỪNG KHÓA')
plt.grid(ls='--')
plt.xlabel('Số lượng')
plt.ylabel('Khóa')

plt.show()
```

```python
#vẽ biểu đồ tròn
plt.pie(so_hs_y,
       labels=ten_khoa_x,
       autopct='%.1f%%',
       wedgeprops=dict(width=0.7, edgecolor='w'))
plt.title('BIỂU ĐỒ THỂ HIỆN TỶ LỆ % SINH VIÊN THEO TỪNG KHÓA')
plt.show()
```

```python
#Hiển thị 2 biểu đồ vào cùng 1 khung
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)

plt.bar(ten_khoa_x, #dữ liệu trục x
        so_hs_y, #dữ liệu trục y
        color='red') #màu biểu đồ
plt.title('BIỂU ĐỒ THỂ HIỆN SỐ LƯỢNG SV THEO TỪNG KHÓA')
plt.grid(ls='--')
plt.xlabel('Khóa')
plt.ylabel('Số lượng')

plt.subplot(1,2,2)
plt.pie(so_hs_y,
       labels=ten_khoa_x,
       autopct='%.1f%%',
       wedgeprops=dict(width=0.7, edgecolor='w'))

plt.show()
```
